### 3. Display the percentage distribution of the data in each year using the pie chart.


```python
# Group the data by year and calculate the sum of each column for each year
yearly_data = df.groupby('year').sum()

# Plot the pie chart
yearly_data.plot.pie(y='totalprod', figsize=(10, 10), autopct='%1.1f%%')
plt.title('Percentage Distribution of Total Honey Production by Year')
plt.ylabel('totalprod')
plt.show()
```

### 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python
sns.displot(data = df, x = df['priceperlb'], kind = 'kde')
plt.title('Distribution of the variable price per lb')
```

### 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
plt.figure(figsize = (25,10))
sns.boxplot(df, x = 'year', y = 'prodvalue')
plt.title('Boxplot of prodvalue by year')
plt.show()
```

- In the year 2010 and 2012 has the high positive outliers belongs to prodvalue.
- This Boxplot represents the no negative outliers present in this variable prodvalue
- Over the year the outliers are minimum in the initial stage and maximum in the end of the period

### 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol','total prod', 'stocks', 'price per lb', 'prodvalue')

```python
cols = [ 'numcol', 'yieldpercol','totalprod', 'stocks', 'priceperlb', 'prodvalue']
cols
sns.heatmap(df[cols].corr(),annot = True)
plt.title('Correlation Heatmap')
plt.show()
```

- And prodvalue with numcol and totalprod has the good relation between them with 91% positively correlated
